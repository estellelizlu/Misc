```python
# Use GroupBy() & compute sum on specific column
df1.groupby('ParentAirline')['Current'].sum() 
df4=df1.sort_values (by = ['Current'], ascending=False)

df4=df4[['ParentAirline', 'Current']].head(10)
df4
```
